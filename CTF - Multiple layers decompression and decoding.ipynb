```python
import zlib
import bz2
import binascii
import lzma
import base64
import gzip

TEMP_MESS = r"C:\Users\ZhenHui\Desktop\Github\CTF-Programming\CSIT CTF 20200808\temp.mess"
TEMP_FILE =  r"C:\Users\ZhenHui\Desktop\Github\CTF-Programming\CSIT CTF 20200808\temp" #Save output to this folder
```

```python
def write_temp(string):
    with open(TEMP_FILE, 'br+') as f:
        f.write(string)
        
# Read mulitple layer encryption folder
def read_tempmess():
    f = open(TEMP_MESS, 'r')
    return f.readline()

def false():
    print("FALSE. Returned previous reading. ", end='\r')

# Check if string is base64
def isBase64(s):
    try:
        return base64.b64encode(base64.b64decode(s)) == s
    except Exception:
        return False

# Decode base64
def decompress_base64(s):
    try:
        check = base64.b64encode(s)
        if isBase64(check) == True:
            compressed = base64.decodebytes(s)
            print("True input is base64. Length: " + str(len(compressed)) + " Output sample: " + str(compressed[0:20]))
            return compressed, True
        else:
            false()
            return s, False
    except Exception:
        false()
        return s, False

# Decode binary
def decompress_binary(s):
    try:
        compressed = binascii.unhexlify(s)
        print("True input is binary. Length: " + str(len(compressed)) + " Output sample: " + str(compressed[0:20]))
        return compressed, True
    except:
        false()
        return s, False

# Decompress bzip2 
def decompress_bz2(s):
    try:
        compressed = bz2.decompress(s)
        print("True input is bz2. Length: " + str(len(compressed)) + " Output sample: " + str(compressed[0:20]))
        if len(compressed) > 10:
            return compressed,True
        else:
            false()
            return s,False
    except:
        false()
        return s,False
    
# Decompress xz 
def decompress_xz(s):
    try:
        write_temp(s)
        with lzma.open(TEMP_FILE) as f:
            compressed = f.read()
            print("True input is xz. Length: " + str(len(compressed)) + " Output sample: " + str(compressed[0:20]))
        return compressed,True
    except:
        false()
        return s,False
    
# Decompress gzip folder
def decompress_gzip(s):
    try:
        #f=gzip.open(temp,'rb')
        #file_content2 = f.read()
        ba = bytearray(s)
        compressed = zlib.decompress(ba, 15+32)
        print("True input is xz. Length: " + str(len(compressed)) + " Output sample: " + str(compressed[0:20]))
        return compressed,True
    except:
        false()
        return s,False
```
